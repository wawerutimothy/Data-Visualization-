```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head(8)
```

```python
sns.displot(crash_df['not_distracted'], bins=10)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

```python
sns.jointplot(x='ins_losses', y='ins_premium', data=crash_df, kind='reg')
```

A KDE plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

A KDE plot on alchol usage

```python
sns.kdeplot(crash_df['alcohol'])
```

A pair plot to compare distribution across the table.

```python
sns.pairplot(crash_df)
```

Load a dataset for iris for pair grid

```python
iris = sns.load_dataset('iris')
iris.head()
```

A pair grid

```python
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map(plt.scatter)
```

Plot a scatter plot and histogram

```python
plt.figure(figsize=(12,6))
sns.set_context('talk', font_scale=0.9)
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
plt.figure(figsize=(12,9))
sns.set_context('talk', font_scale=0.9)
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

A custom pair grid

```python
plt.figure(figsize=(12,6))
sns.set_context('talk', font_scale=0.9)
iris_g = sns.PairGrid(iris, hue='species', x_vars=['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

A facet grid

```python
plt.figure(figsize=(25,15))
sns.set_context(font_scale=0.8)
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill',  bins=8)
```

```python
sns.set_style('white')
plt.figure(figsize=(25,15))
sns.set_context('talk', font_scale=0.8)
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.scatter, 'total_bill',  'tip')
```

To add variable to color

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3, col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```

Additional Facet grid functionalities

```python
kws = dict(s=50, linewidth=.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3, hue_order=['Yes', 'No'], 
                        hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```
